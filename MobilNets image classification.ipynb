```python
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.layers import Dense , Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import CategoricalCrossentropy
from tensorflow.keras.preprocessing import image
```

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
import os
os.chdir('D:\cvdata')
def prepare_image(file):
    img_path = 'D:\cvdata'
    img = image.load_img(img_path + file ,target_size=(224,224))
    img_array = image.img_to_array(img)
    img_expand = np.expand_dims(img_array,axis = 0)
    return tf.keras.applications.mobilenet.preprocess_input(img_expand)
```

```python
from IPython.display import Image
Image(filename='dog1.jpeg',width=450,height=300)
```

```python
preprocessed_image = prepare_image('\dog1.jpeg')
predict = model.predict(preprocessed_image)
```

```python
predict
```

```python
from tensorflow.keras.applications import imagenet_utils
result = imagenet_utils.decode_predictions(predict)
```
